```python
#1 train test set
from sklearn.model_selection import train_test_split
```

```python
x,y=h.drop('target',axis=1),h['target']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#2 
# Training data is used to learn patterns from data
# Testing data checks how well the model performs on unseen data
# This avoids overfitting and ensures real-world reliability
```

```python
#3 model logistic regression
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(x_train, y_train)
```

```python
model.fit(x_train,y_train)
```

```python
#4 predict the test
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
#5
from sklearn.metrics import accuracy_score, precision_score, recall_score
```

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
```

```python
#6 confusion matrix
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
#7 interpret results
# High accuracy → Model performs well overall
# High precision → Few false positives
# High recall → Few missed positive cases
# Balanced scores indicate a reliable classification model
```
